```python
import numpy as np
from matplotlib import pyplot as plt

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)

plt.title("Sample Visualization")
plt.show()
```

Cosa significano le variabili vs e x?

```python
print("ys=", ys)
print("x=", x)
```

Quindi:
- ys sono valori 100 valori attorno a 200

```python
print("Sequenza di valori della distribuzione gaussiana:", ys-200)
```

Procediamo con l'esercizio:
- modificare lo script per avere solo 10 valori di x

```python
ys1 = 200 + np.random.randn(10)
x1 = [x1 for x1 in range(len(ys1))] 
print("ys1=", ys1)
print("x1=", x1)
```

- Moltiplicare per 10 la variazione casuale di ys

```python
ys1 = 200 + 10*np.random.randn(10)
x1 = [x1 for x1 in range(len(ys1))] 
print("ys1=", ys1)
print("x1=", x1)
```

- Visualizzo il grafico dopo questi cambiamenti:

```python
plt.plot(x1, ys1, '-')
plt.fill_between(x1, ys1, 195, where=(ys1 > 195), facecolor='yellow', alpha=0.5)

plt.title("Nuova Visualizzazione")
plt.show()
```

Il grafico non è molto uniforme quindi provo a cambiare i valori di fill_between in modo tale che i valori minimi x1 e ys1 colorino il grafico:

```python
plt.plot(x1, ys1, '-')
plt.fill_between(x1, ys1, 160, where=(ys1 > 160), facecolor='yellow', alpha=0.5)

plt.title("Nuova Visualizzazione")
plt.show()
```
