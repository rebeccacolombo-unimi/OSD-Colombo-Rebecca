## ESEMPI DI GRAFICI 

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
sns.set_theme(style="darkgrid")
```

```python
tips=sns.load_dataset("tips")
tips
```

```python
tips=tips.rename(columns={"total_bill":"bill"})
tips
```

```python
uomini= (tips.sex =="Male")
donne= (tips.sex== "Female")
```

```python
tips_uomini=tips[uomini]
tips_uomini
```

```python
tips_donne=tips[donne]
tips_donne
```

**Creo un grafico in cui pongo a confronto la variabile indipendente "bill" quindi le fatture e le variabili dipendenti "tip" con maschi e femmine divisi**

```python
sns.scatterplot(data=tips_uomini, x="bill", y="tip", color="blue");
sns.scatterplot(data=tips_donne, x="bill", y="tip", color="red");
plt.show()
```

```python
sns.relplot(data=tips_uomini, x="bill", y="tip", kind="scatter", color="blue");
sns.relplot(data=tips_donne, x="bill", y="tip", kind="scatter", color="red");
plt.show()
```

```python
sns.relplot(data=tips_uomini, x="bill", y="tip", kind="line", color="blue");
sns.relplot(data=tips_donne, x="bill", y="tip", kind="line", color="red");
plt.show()
```

```python
sns.lineplot(data=tips_uomini, x="bill", y="tip", color="blue");
sns.lineplot(data=tips_donne, x="bill", y="tip", color="red");
plt.show()
```
