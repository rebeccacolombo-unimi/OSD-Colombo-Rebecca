```python
import numpy as np
import pandas as pd
```

```python
data1 = pd.read_sas('../data/LLCP2018.XPT')
data1
```

```python
peso_alt1 = pd.DataFrame(data1, columns=['SEX1', '_STATE', 'HTM4', 'WTKG3'])
peso_alt1['WTKG3']=peso_alt1['WTKG3']/100
print(peso_alt1)
```

```python
peso_alt1.iloc[:,0]=peso_alt1.iloc[:,0].astype('str') 
peso_alt1.iloc[:,0]=peso_alt1.iloc[:,0].replace('2.0', "female")
peso_alt1.iloc[:,0]=peso_alt1.iloc[:,0].replace('1.0', "male")
peso_alt1.iloc[:,0]=peso_alt1.iloc[:,0].replace('7.0', "don't know/not sure")
peso_alt1.iloc[:,0]=peso_alt1.iloc[:,0].replace('9.0', "refused")
peso_alt = peso_alt1.rename(columns={"SEX1":"Gender", "_STATE":"Cod_Stato", "HTM4": "Height", "WTKG3": "Weight"})
print(peso_alt)
```

```python
peso_alt.groupby(['Cod_Stato','Gender'])[['Height']].mean()
```

```python
peso_alt.groupby(['Cod_Stato','Gender'])[['Height']].aggregate('mean')
```

```python
peso_alt.groupby(['Cod_Stato','Gender'])[['Height']].aggregate('mean').unstack()
```

```python
peso_alt.groupby(['Cod_Stato','Gender'])[['Height']].aggregate(['min', np.median, max]).unstack()
```

Ora con PIVOT TABLE (metodo di aggregazione di default: media)


```python
peso_alt.pivot_table('Height', index='Cod_Stato', columns='Gender')
```

```python
peso_alt.pivot_table(index='Cod_Stato', columns='Gender', aggfunc={'Height':[max,np.median,min]})
```

```python
peso_alt.pivot_table(index='Cod_Stato', columns='Gender', aggfunc={'Height':[max,min], 'Weight':'mean'})
```

**SELEZIONIAMO SOLO QUELLI PIÙ ALTI DI 200 CM**

```python
peso_alt1 = peso_alt.query('(Height > 200)')
peso_alt1
```

```python
peso_alt1.pivot_table(index='Cod_Stato', columns='Gender')
```

```python
peso_alt1.pivot_table(index='Cod_Stato', columns='Gender', aggfunc={'Weight':[max,'mean',min]})
```
