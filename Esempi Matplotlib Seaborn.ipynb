**Importo il dataset "tips" dallo ZIP su gytub:https://github.com/mwaskom/seaborn-data**

```python
tips = pd.read_csv("C:/Users/utente/Desktop/UNIMI/Organizzazioni e società digitali/Esercizi/tips.csv")
tips
#oppure posso fare: tips =sns.load_dataset("tips")
#tips
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tips) #posso utilizzare anche sns.relplot che è più in generale e successivamente scelgo il tipo di grafico mettendo ad esempio (x="", y="", kind="scatter")
plt.show()
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip",hue="smoker") #oppure hue="sex" per vedere donne e uomini
plt.show()
```

```python
#cambiamo ordine smoker e colore pallini
hue_colors = {"Yes":"blue", "No": "red"}
sns.scatterplot(data=tips, x="total_bill", y="tip",hue="smoker",  hue_order=["Yes", "No"], palette=hue_colors) 
plt.show()
```

```python
#oppure sempre per cambiare il colore scelgo il codice RGB (vedere su internet tabella colori):
```

```python
hue_colors = {"Yes":"#7FFFD4", "No": "#007FFF"}
sns.scatterplot(data=tips, x="total_bill", y="tip",hue="smoker",  hue_order=["Yes", "No"], palette=hue_colors, 
                size="tip", sizes=(20, 100)) 
plt.show()
```

Importo un altro dataset

```python
fl = sns.load_dataset("flights")
fl
```

```python
sns.relplot(data=fl, x="year", y="passengers", kind="line", hue="month", palette="mako") #oppure kind="scatter"
plt.show()
#posso anche usare sns.scatterplot 
```

```python
sns.relplot(data=fl,x="year", y="passengers",kind="scatter", hue="month", palette="mako", col="month", col_wrap=4)#col_wrap dice vai a capo dopo 4 grafici
plt.show()
#questo grafico in realtà non mi dice molto
```
